# eval() and query()  LESS COMPUTATION TIME EFFICIENT MEMORY USAGE

```python
import numpy as np
import pandas as pd
rng=np.random.RandomState(42)
nrows,ncols=1000,10
df1,df2,df3,df4=(pd.DataFrame(rng.rand(nrows,ncols)) for i in range(4))
```

```python
%timeit df1+df2+df3+df4
```

```python
%timeit pd.eval('df1+df2+df3+df4')
```

```python
r1=df1+df2+df3+df4
r2=pd.eval('df1+df2+df3+df4')
np.allclose(r1,r2)
```

```python
#we can use arithmetic,comparision,bitwise,etc operations on dataframe using eval() method
```

```python
r1=(df1+df2)/(df1-df2)**2+df3
r2=pd.eval('(df1+df2)/(df1-df2)**2+df3')
np.allclose(r1,r2)
```

```python
#assignment using df.eval() method
df1.eval('C=0',inplace=True)
df1
```

```python
#query() method is used to filter out data in a dataframe
```

```python
rng=np.random.RandomState(0)
nrows,ncols=1000,10
df=pd.DataFrame({'A':rng.randn(nrows),'B':rng.randn(nrows)})
df
```

```python
df.eval('C=(A+B)/2') 
```

```python
#selecting data using query()
r1=df[df['A']<2]
r2=df.query('A<2')

%timeit r1
```

```python
%timeit r2
```
