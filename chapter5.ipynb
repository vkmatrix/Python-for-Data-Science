# Line Plot

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,marker='x',markersize=10,markeredgewidth=4,markeredgecolor='green',markerfacecolor='orange')
```

```python
plt.plot(x,x**3,linestyle='--',color='violet')
```

```python
plt.plot(x,y,'o')
plt.plot(x,np.cos(x),'v')
plt.plot(x,x-4,'--')
```

# Scatter Plot

```python
plt.style.use('seaborn-whitegrid')
```

```python
x=np.linspace(0,10,60)
y=np.sin(x)
plt.plot(x,y,marker='o',color='green',label='H',markerfacecolor='red',markeredgecolor='blue',markersize=8,markeredgewidth=2,linewidth=4)
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.title('SINE GRAPH')
plt.legend(frameon=False,loc='lower right');
```

# 3D Plotting 

```python
from mpl_toolkits import mplot3d
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot3D(x,y,np.tan(x),'gray');
```

```python
xl=np.linspace(0,15,100)
yl=np.cos(xl)
zl=np.sin(xl)
ax=plt.axes(projection='3d')
ax.plot3D(xl,yl,zl,'--r')
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,85,cmap='viridis');
```
